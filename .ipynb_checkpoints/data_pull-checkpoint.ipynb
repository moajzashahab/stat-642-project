```python
import time
import fastf1
import logging
import warnings
import pandas as pd
from datetime import datetime

# enable cache
fastf1.Cache.enable_cache("cache")
```

```python
all_schedules = []

for year in range(2018, 2027):  # 2018 through 2026
    try:
        schedule = fastf1.get_event_schedule(year)

        # write one file per year
        schedule.to_csv(f"data/event_schedule_{year}.csv", index=False)

        # optional: also keep one combined file later
        temp = schedule.copy()
        temp["Season"] = year
        all_schedules.append(temp)

        print(f"saved {year}")
    except Exception as e:
        print(f"skipping {year}: {e}")

# optional combined file
if all_schedules:
    full_schedule = pd.concat(all_schedules, ignore_index=True)
    full_schedule.to_csv("data/event_schedule.csv", index=False)
```
